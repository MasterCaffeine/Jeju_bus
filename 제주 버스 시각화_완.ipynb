```r
ride = train[, c('date','X6.7_ride', 'X7.8_ride', 'X8.9_ride', 'X9.10_ride','X10.11_ride' , 'X11.12_ride','X18.20_ride')]
take_off = train[, c('date','X6.7_takeoff', 'X7.8_takeoff', 'X8.9_takeoff', 'X9.10_takeoff', 'X10.11_takeoff', 'X11.12_takeoff','X18.20_ride')]
```

```r
A = filter(ride, X18.20_ride==0)
A['rowsum'] = rowSums(filter(ride[,2:8], X18.20_ride==0))
B = filter(take_off, X18.20_ride == 0)
B['rowsum'] = rowSums(filter(take_off[,2:8], X18.20_ride==0))
AB = filter(train[,9:21], X18.20_ride ==0)
AB['rowsum'] = rowSums(filter(train[,9:21], X18.20_ride==0))
```

```r
# Train의 승차 칼럼들의 rowsum
max(A['rowsum'])
min(A['rowsum'])
# Train의 하차 칼럼들의 rowsum
max(B['rowsum'])
min(B['rowsum'])
# Train의 승하차 합 칼럼들의 rowsum
max(AB['rowsum'])
min(AB['rowsum'])
```

```r
bus_time = train[,c('X6.7_ride', 'X7.8_ride', 'X8.9_ride', 'X9.10_ride', 'X11.12_ride',
             'X6.7_takeoff', 'X7.8_takeoff', 'X8.9_takeoff', 'X9.10_takeoff', 'X11.12_takeoff',
             'X18.20_ride')]

HM_2 = cor(bus_time)

corrplot(HM_2,
        method = 'color',
        type = 'lower',
        order = 'hclust',
        tl.col = 'black',
         tl.srt=45,
        addCoef.col = 'white')
```

```r
train['68a']=train['X6.7_ride']+train['X7.8_ride'] # 6 ~ 8시 승차인원 (2시간단위로 묶어줌)
train['810a']=train['X8.9_ride']+train['X9.10_ride']
train['1012a']=train['X10.11_ride']+train['X11.12_ride']

train['68b']=train['X6.7_takeoff']+train['X7.8_takeoff'] # 6 ~ 8시 하차인원
train['810b']=train['X8.9_takeoff']+train['X9.10_takeoff']
train['1012b']=train['X10.11_takeoff']+train['X11.12_takeoff']

HM = train[,c('68a','810a','1012a','68b','810b','1012b', 'X18.20_ride')]
```
